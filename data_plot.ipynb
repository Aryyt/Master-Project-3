```python
def split_array_and_save_csv(
    arr: np.ndarray,
    n: int,
    out_dir: str,
    base_name: str = "subset",
    shuffle: bool = False,
    random_state: Optional[int] = None,
    delimiter: str = ",",
    fmt: Union[str, List[str]] = "%s",
) -> List[Path]:
    """
    Split a NumPy array into n roughly equal parts and save each to CSV.

    Args:
        arr: Input NumPy array. If 1D, each chunk is saved as a single-column CSV.
        n: Number of subsets (>=1).
        out_dir: Directory to write CSV files (created if it doesn't exist).
        base_name: Base filename; files will be base_name_1..base_name_n.csv.
        shuffle: Randomly permute rows before splitting (for 2D, shuffles along axis=0).
        random_state: Seed used when shuffle=True.
        delimiter: Delimiter for CSV files.
        fmt: Format string(s) for np.savetxt.
    Returns:
        List of Path objects to the saved CSV files.
    """
    if n < 1:
        raise ValueError("n must be >= 1")

    out_path = Path(out_dir)
    out_path.mkdir(parents=True, exist_ok=True)

    arr = np.asarray(arr)
    if arr.ndim == 1:
        arr2d = arr.reshape(-1, 1)
    elif arr.ndim == 2:
        arr2d = arr
    else:
        raise ValueError("Only 1D or 2D arrays are supported")

    total = arr2d.shape[0]

    # Optionally shuffle rows
    if shuffle and total > 0:
        rng = np.random.default_rng(random_state)
        idx = rng.permutation(total)
        arr2d = arr2d[idx]

    # Prepare even chunk sizes
    base_chunk = total // n
    remainder = total % n

    paths: List[Path] = []
    start = 0
    for i in range(1, n + 1):
        size = base_chunk + (1 if i <= remainder else 0)
        end = start + size
        chunk = arr2d[start:end]

        file_path = out_path / f"{base_name}_{i}.csv"
        # If chunk is empty, still create an empty file
        if chunk.size == 0:
            file_path.touch()
        else:
            np.savetxt(file_path, chunk, delimiter=delimiter, fmt=fmt)
        paths.append(file_path)
        start = end

    return paths
```
